```python
#task 3
n=5
arr=[]
sum=0;

i=0
while(i!=n):
  arr.append(int(input("Enter a Number: ")))
  i+=1

for i in arr:
  sum+=i;
average=sum/n;

print("sum is: ",sum);
print("average is: ",average)

```

```python
#Task 4
attended=int(input("Enter no of classes attended: "))
held=int(input("Enter no classes held: "))
percent=attended*100/held;
print("Student percentage is: ",percent)
if(percent<75):
  print("Student is not allowed to attend exam!")
else:
  print("Student is allowed to attend exam!")
```
