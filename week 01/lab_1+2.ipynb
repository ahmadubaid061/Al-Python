```python
#task 5
physicMarks=int(input("Enter obtained marks in physics: "))
chemistryMarks=int(input("Enter obtained marks in chemistry: "))
biologyMarks=int(input("Enter obtained marks in Biology: "))
mathMarks=int(input("Enter obtained marks in mathematics: "))
computermarks=int(input("Enter obtained marks in Computer: "))

def PrintGrade(x):
  if(x>=90):
    print("A")
  elif(x>=80):
    print("B")
  elif(x>=70):
    print("C")
  elif(x>=60):
    print("D")
  elif(x>=40):
    print("E")
  elif(x<40):
    print("F")
print("Grade in physics is: ",end="")
PrintGrade(physicMarks)
print("Grade in chemistry is: ",end="")
PrintGrade(chemistryMarks)
print("Grade in Biology is: ",end="")
PrintGrade(biologyMarks)
print("Grade in Mathematics is: ",end="")
PrintGrade(mathMarks)
print("Grade in Computer is: ",end="")
PrintGrade(computermarks)
```

```python
#task 10
digits=0
letters=0
mystring=input("Enter a string: ")
for char in mystring:
  if(char.isalpha()):
    letters+=1
  elif(char.isdigit()):
    digits+=1
print(f"count of Digits is: {digits}")
print(f"count of Letters is: {letters}")

```

```python
#task 11
password=input("Enter your password: ")
if(len(password)<6):
  print("Error: Password must be atleast 6 charactor long!")
elif(len(password)>16):
  print("Error! password cannot exceed 16 charactors!")

digits=0;
special=0;
capital=0;
small=0;
for char in password:
  if(char.isupper()):
    capital+=1
  elif(char.islower()):
    small+=1
  elif(char.isdigit()):
    digits+=1;
  elif(not char.isspace()):
    special+=1;

if(digits==0 or small==0 or capital==0 or special==0):
  print("Error: Password must contain atleast one letter(a to z) and one letter (A to Z), atleast a digit (0 to 9) and atleast a special charactor!")
```

```python
#task 12
n = 9
mylist = []
for i in range(n):
    value = int(input(f"Enter value {i}: "))
    mylist.append(value)
num = 2
while len(mylist) > 0 and num <= len(mylist):
    print(f"Number deleted from index {num} is: {mylist[num]}")
    del mylist[num]
    print("original list is now: ",end='')
    for i in mylist:
      print(i,end=',')
    print()
    if len(mylist) == 0:
        break
    num = (num + 3) % len(mylist)
```
