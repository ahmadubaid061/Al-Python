Task #2:
Create a simple book inventory system using tuples. Each book in the inventory should be represented as a tuple containing the following information:
- Title
- Author
- Publication Year
- ISBN (International Standard Book Number)
- Quantity in Stock

1. Define an empty list named book_inventory to store tuples representing books.
2. Implement a function add_book that takes input for a new book and adds it to the book_inventory. The function should prompt the user for the book details (title, author, publication year, ISBN, quantity) and create a tuple to represent the book. Append this tuple to the book_inventory.
3. Implement a function search_book that takes the ISBN as input and searches for the book in the inventory. If the book is found, print its details; otherwise, print a message indicating that the book is not in the inventory.
4. Implement a function display_inventory that prints the details of all books in the inventory. Each book's information should be printed on a new line.
5. Implement a function sell_book that takes the ISBN and quantity as input and updates the stock quantity of the corresponding book in the inventory. If the quantity is greater than the available stock, print a message indicating insufficient stock.
6. Test your inventory system by adding a few books, searching for them, displaying the inventory, and selling books.

```python
book_inventory=[]
def add_book():
  title=input("Enter title of the book : ")
  author=input("Enter author of the book : ")
  publication_year=input("Enter year of publication : ")
  ISBN=input("Enter ISBN : ")
  quantity=input("Enter quantity of the book : ")
  book=(title, author, publication_year, ISBN, quantity)
  book_inventory.append(book)

```

displaying inventory funciton

```python
def display_inventory():
   if not book_inventory:
        print("Inventory is empty.")
   else:
        print("Printing inventory details!")
        print("{",end='')
        for book in book_inventory:

            print(f" Title: {book[0]}, Author: {book[1]}, Year: {book[2]}, ISBN: {book[3]}, Quantity: {book[4]}")
        print(" }")
```

funciton for searching by ISBN

```python
def search_book(isbn):
  for book in book_inventory:
    found=False
    if str(isbn) == str(book[3]):
      found=True
      print(f" Book found:: Title : {book[0]}, Author: {book[1]}, Publication_Year : {book[2]}, ISBN : {book[3]}, Quantity : {book[4]}")
      break
  if found==False:
      print("Book not found!")
```

```python
def sell_book():
  isbn=input("Enter ISBN of the book you want to sell : ")
  quantity=int(input("Enter quantity you want to sell : "))
  found=False
  for i,book in enumerate(book_inventory):
    if(isbn==book[3]):
      found=True
      if(int(book[4])<quantity):
        print("That quantity of book is not availible in stock!")
      else:
        new_quantity=int(book[4])-quantity
        updated_book=(book[0],book[1],book[2],book[3],new_quantity)
        book_inventory[i]=updated_book
        print(f"{quantity} books sold! remaining quantity of the book is {new_quantity}")
    if found==False:
      print("book not found!")
```

testing the book_inventory

```python
print("adding a book")
add_book()
print("adding another book")
add_book()
display_inventory()
search_book(4095)
sell_book()
display_inventory()
```
