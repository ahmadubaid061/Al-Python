**BFS**

```python
myGraph={
    '1':['2','3','4'],
    '2':['5','6'],
    '3':[],
    '4':['7','8'],
    '5':['9','10'],
    '6':[],
    '7':['11','12'],
    '8':[],
    '9':[],
    '10':[],
    '11':[],
    '12':[]
}

def bfs(visited, graph, startingNode,goal):
#by adding the fourth perimeter'goal' this funciton can be reused any where
    queue=[]

    visited.append(startingNode)
    queue.append(startingNode)

    while queue:
        key = queue.pop(0)

        if key == goal:
            print(key)
            break
        #else
        print(key, end=' ')

        for vals in graph[key]:
            if vals not in visited:
                visited.append(vals)
                queue.append(vals)


visitedlist=[]
bfs(visitedlist,myGraph,'1','11')
```

**DFS**

```python
def dfs(visited, graph, startingNode,goal):
    stack=[]

    if startingNode not in visited:
        visited.append(startingNode)
        stack.append(startingNode)

    while stack:
        key = stack.pop()
        print(key, end=' ')

        if key == goal:
            break

        for vals in reversed(graph[key]):
            # reversed() insures the smaller values be on top
            if vals not in visited:
                visited.append(vals)
                stack.append(vals)


visitednewList=[]
dfs(visitednewList, myGraph, '1','11')
```

**GRPAH**

```python
graph3={
    'A':[('B',5),('C',7)],
    'B':[('D',10),('E',10)],
    'C':[('F',8)],
    'D':[],
    'E':[],
    'F':[]
}
```

**BFS**

```python
def bfs2(visited, graph, start, goal):

    queue=[]
    queue.append((start,0))   # node , cost
    visited.append(start)

    while queue:

        node, cost = queue.pop(0)

        print(node,end=' ')

        if node == goal:
            print("\nTotal Cost:",cost)
            break

        for neighbor, weight in graph[node]:
            if neighbor not in visited:
                visited.append(neighbor)
                queue.append((neighbor, cost + weight))


visitedlist=[]
bfs2(visitedlist,graph3,'A','F') #starting node A and goal F
```

**DFS**

```python
def dfs2(visited,graph,start,goal):

    stack=[]
    stack.append((start,0))

    while stack:

        node,cost=stack.pop()

        if node not in visited:

            visited.append(node)
            print(node,end=' ')

            if node==goal:
                print("\nTotal Cost:",cost)
                break

            for neighbor,weight in graph[node]:
                stack.append((neighbor,cost+weight))


visited_list=[]
dfs2(visited_list,graph3,'A','F')
```
