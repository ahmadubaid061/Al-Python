Make a class parallelogram and write a function to calculate its area.


```python
class Parallelogram(object):
  def __init__(self,y,z):
    self.base=y
    self.height=z

  def calArea(self):
    self.area=self.base*self.height
```

creating object of class parallelogram

```python
parrallelogram=Parallelogram(4,5)
parrallelogram.calArea()
print("Area of parrallelogram is: ",parrallelogram.area)
```
