```python
rooms={}
location=input("Enter current Location(A,B,C,D: ")
rooms['A']=input("Enter status of room A(0/1): ")
rooms['B']=input("Enter status of room B(0/1): ")
rooms['C']=input("Enter status of room C(0/1): ")
rooms['D']=input("Enter status of room D(0/1): ")

def cleaner(dictionary,current):
    path_cost=0
    path_order=['A', 'B', 'C','D']
    starting_node=path_order.index(current)
    for i in range(starting_node,len(path_order)):
        current_room=path_order[i]
        print(f"Agent is in room {current_room}")
        
        if dictionary[current_room] == '0':
            print(f"Room {current_room} is already cleaned")
        
        elif dictionary[current_room] == '1':
            print(f"Cleaning room {current_room}....")
            dictionary[current_room] = '0'
        
        path_cost += 1

        if current_room != 'D':
           print(f"Moving toward room {path_order[i+1]}....")
    print(f"total Path cost is: {path_cost}")

cleaner(rooms,location)
            
```

```python
def guideMe(position):
    match position:
        case 'S':
            print("move toward position A")
        case 'A':
            print("move toward position B")
        case 'B':
            print("move toward position C")
        case 'C':
            print("move toward position G")
        case 'G':
            print("You are Already at target position!")
        case _:
            print("Invalid choice")
currentPosition=input("Enter the current position: ")
guideMe(currentPosition)
```
