```python
graph = {
    'A': {'B': 1, 'C': 4},
    'B': {'D': 3, 'C': 2},
    'C': {'E': 5},
    'D': {'F': 2, 'G': 4},
    'E': {'G': 3},
    'F': {'G': 1},
    'G': {}
}
h = {
    'A': 5,
    'B': 6,
    'C': 4,
    'D': 3,
    'E': 3,
    'F': 1,
    'G': 0
}
def A_star_search(start, goal):
    open_list = [start]
    g_cost = {start: 0}  #actual cost from start node to current node
    parent = {start: None}  #stores parent nodde through which the current node is explored
                            #parent = {start: None} means starting node has no parent node
#=======================================================================================
    while open_list:
        current = min(open_list, key=lambda x: g_cost[x] + h[x]) #stores the node with minimum cost(f(n)) in current
        if current == goal:
            path = []
            while current is not None:
                path.append(current)                #add current node to path
                current = parent[current]           #current: the parent of current is now current node
            path.reverse()
            print("Path found:", " → ".join(path))  #.join converts list into a string
            print("Total cost:", g_cost[goal])
            return
        open_list.remove(current)
        for neighbor, cost in graph[current].items():
            temp_g = g_cost[current] + cost
            if neighbor not in g_cost or temp_g < g_cost[neighbor]:
                g_cost[neighbor] = temp_g
                parent[neighbor] = current
                if neighbor not in open_list:
                    open_list.append(neighbor)
#==========================================================================================
    print("No path found!")
A_star_search('A', 'G')
```
