```python
import math
board = [" "] * 9
def draw_board():
    print()
    print(f" {board[0]} | {board[1]} | {board[2]} ")
    print("---+---+---")
    print(f" {board[3]} | {board[4]} | {board[5]} ")
    print("---+---+---")
    print(f" {board[6]} | {board[7]} | {board[8]} ")
    print()

def check_winner():
    wins = [
        [0,1,2], [3,4,5], [6,7,8],  # rows
        [0,3,6], [1,4,7], [2,5,8],  # columns
        [0,4,8], [2,4,6]             # diagonals
    ]
    for a, b, c in wins:
        if board[a] == board[b] == board[c] and board[a] != " ":
            return board[a]  # "X" or "O"

    if " " not in board:
        return "tie"

    return None  # game still going


def minimax(is_maximizing):
    result = check_winner()

    
    if result == "O":  
        return 1
    if result == "X":  
        return -1
    if result == "tie":
        return 0

    if is_maximizing:
        
        best_score = -math.inf 
        for i in range(9):
            if board[i] == " ":          
                board[i] = "O"           
                score = minimax(False)    
                board[i] = " "           
                best_score = max(best_score, score)  

        return best_score

    else:
      
        best_score = math.inf  

        for i in range(9):
            if board[i] == " ":           
                board[i] = "X"           
                score = minimax(True)     
                board[i] = " "           
                best_score = min(best_score, score)  

        return best_score


def computer_move():
    best_score = -math.inf
    best_spot = None

    for i in range(9):
        if board[i] == " ":           
            board[i] = "O"            
            score = minimax(False)    
            board[i] = " "            
            if score > best_score:    
                best_score = score
                best_spot = i         

    board[best_spot] = "O"           
    print(f"Computer placed O at position {best_spot + 1}")



def human_move():
    while True:
        print("Your turn (X) - enter a position (1-9): ", end="")
        move = int(input()) - 1      

        if 0 <= move <= 8 and board[move] == " ":
            board[move] = "X"
            break
        else:
            print("Invalid move! Try again.")


print("=== Tic Tac Toe — You vs Computer ===")
print("Positions:")
print(" 1 | 2 | 3 ")
print("---+---+---")
print(" 4 | 5 | 6 ")
print("---+---+---")
print(" 7 | 8 | 9 ")
print()
print("You are X. Computer is O.")

while True:
    draw_board()

    
    human_move()
    result = check_winner()
    if result is not None:
        draw_board()
        if result == "tie":
            print("It's a tie!")
        else:
            print("You win! (Impressive — the computer never loses normally!)")
        break

    
    computer_move()
    result = check_winner()
    if result is not None:
        draw_board()
        if result == "tie":
            print("It's a tie!")
        else:
            print("Computer wins!")
        break
```
